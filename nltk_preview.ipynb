## 2. 计算特朗普的总统任职演讲

* 读取
* 分词
* 制作自己的nltk text对象
* 绘制关键词语分布图
* 查找最热的长度为2的短语
* 计算词频
* 频率最高的词语或符号
* 累计统计图
* 用词多样化的程度


````python
with open('./data/trump.txt','r') as f:
    trump = f.read()
````

````python
trump
````

### 2.1 分词

tokens = nltk.word_tokenize(trump) #取出一句话中的单词

text = nltk.Text(tokens) # 创建nltk的文本对象，从而可以使用相关函数

````python
tokens = nltk.word_tokenize(trump)
````

````python
tokens
````

````python
text = nltk.Text(tokens)
````

### 2.2 绘制关键词语分布图

text.dispersion_plot(['civilian','people','American'])

````python
text.dispersion_plot(['citizens','democracy','freedom','duties','America'])
text.dispersion_plot(['civilian','people','American'])
text.dispersion_plot(['United','American','Korea'])
````

### 2.3 查找最热的长度为2的短语

text.collocations() #collocations同时存在

````python
text.collocations()
````

### 2.4 计算词频

text_fq = nltk.FreqDist(text)

text_fq.most_common(20) # 가장 많이 나오는 단어


````python
text_fq = nltk.FreqDist(text)
text_fq.most_common(20)
````

### 2.5 词汇累计统计图

text_fq.plot(30,cumulative=True) #cumulative累积的; 渐增的; 追加的


````python
text_fq.plot(30,cumulative=True)
````
